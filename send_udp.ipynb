```python
import socket

STM32_IP = "192.168.1.10"
UDP_CMD_PORT = 55152

def send_cmd(cmd: str) -> None:
    """Envoie une commande texte au STM32 sur le port des commandes UDP."""
    with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as s:
        s.sendto(cmd.encode("utf-8"), (STM32_IP, UDP_CMD_PORT))

# Exemples d'utilisation :
# send_cmd("START")   # Lancer l'acquisition
# send_cmd("STOP")    # Mettre en pause
send_cmd("R_ON")    # LED verte ON
# send_cmd("G_OFF")   # LED verte OFF
# send_cmd("R_ON")    # LED rouge ON
# send_cmd("R_OFF")   # LED rouge OFF
# send_cmd("Y_ON")    # LED jaune ON
# send_cmd("Y_OFF")   # LED jaune OFF
```

```python
import time
for i in range(10):
    send_cmd("G_ON")
    time.sleep(0.01)
    send_cmd("G_OFF")
    time.sleep(0.01)

    send_cmd("R_ON")
    time.sleep(0.01)
    send_cmd("R_OFF")
    time.sleep(0.01)

    send_cmd("Y_ON")
    time.sleep(0.01)
    send_cmd("Y_OFF")
    time.sleep(0.01)
```
